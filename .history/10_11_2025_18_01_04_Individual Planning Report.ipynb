## (1) DATA DESCRIPTION   
**Context**: A research group in UBC computer science, led by Frank Wood, is collecting data about how people play video games. They set up a Minecraft server, and they need to target their recruitment efforts.

The data set "players.csv" has 196 players and 7 variables, and contains information about players who participated in UBC's Minecraft Research Project. Each row represents an individual player. 

**Column Description**  
1. **experience**: The player's experience level in Minecraft (type: character) 
2. **subscribe**: Indicates whether or not the player subscribed to the game (type: logical)
3. **hashedEmail**: The player's anonymous identity (type: character)
4. **played_hours**: The number of hours the player spent playing on the server (type: double)
5. **name**: The player's name (type: character)
6. **gender**: The player's gender (type: character)
7. **Age**: The player's age (type: double)

The variable **played_hours** is our response variable. It represents the number of hours a player has spent on the Minecraft server. By examining other variables, we can gain insights into the type of players that would contribute to the data set the most.

## (2) QUESTIONS  
**Broad question:**  
Which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts?

**Specific question:**   
Can a player's experience level (**experience**), subscription status (**subscribe**), gender (**gender**), and age (**Age**) predict the number of hours players would play on the server? 


We won't be using the columns "hashedEmail" and "name" as they don't provide useful information for the prediction.

## (3) EXPLORATORY DATA ANALYSIS AND VISUALIZATION  
We can further explore the data by doing statistical analysis and creating visualizations.

```r
players |> count(experience)
players |> count(subscribe)
players |> count(gender)
```

The three tables above show the distribution of the categorical variables in the data set.

```r
players_summary <- players |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE))
players_summary

players_summary_based_on_age <- players |>
    group_by(Age) |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE))
players_summary_based_on_age

players_summary_based_on_subscription <- players |>
    group_by(subscribe) |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE))
players_summary_based_on_subscription

players_summary_based_on_gender <- players |>
    group_by(gender) |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE))
players_summary_based_on_gender

players_summary_based_on_experience <- players |>
    group_by(experience) |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE))
players_summary_based_on_experience
```

An issue in the data would be the fact that there is an N/A in the column Age.

```r
age_plot <- ggplot(players_summary_based_on_age, aes(x = Age, y = mean_played_hours)) +
    geom_point() +
    labs(title = "Average Played Hours by Age",
         x = "Player's Age",
         y = "Average Played Hours")
age_plot
```

- There is no clear relationship between the player's age and the average number of hours played on the server.  
- There is no positive or negative relationship between the two variables.
- The player's age may not be a good predictor for predicting the number of hours players play on the server.

```r
subscription_plot <- ggplot(players_summary_based_on_subscription, aes(x = subscribe, y = mean_played_hours, fill = subscribe)) +
    geom_bar(stat = "identity") +
    labs(title = "Average Played Hours by Players' Subscription Status",
         x = "Subscribe?",
         y = "Average Played Hours")
subscription_plot
```

```r
gender_plot <- ggplot(players_summary_based_on_gender, aes(x = gender, y = mean_played_hours, fill = gender)) +
    geom_bar(stat = "identity") +
    labs(title = "Average Played Hours by Gender",
         x = "Gender",
         y = "Average Played Hours")
gender_plot
```

- Players who identify themselves as non-binary have the highest average number of hours of playing in the server, followed by female players.
- Two-spirited-gendered players have the lowest average number of hours playing on the server.

```r
experience_plot <- ggplot(players_summary_based_on_experience, aes(x = experience, y = mean_played_hours, fill = experience)) +
            geom_bar(stat = "identity") +
            labs(title = "Average Played Hours by Experience Level",
            x = "Experience Level",
            y = "Average Played Hours") 
experience_plot
               
```

- Players with the a "Regular" experience level have the highest number of hours of playing in the server.
- This suggest that players who play regularly are more active in the server compared to amateur, beginner, pro, and veteran players.

## (4) METHODS AND PLAN 
- To predict the "kinds" of players who would contribute a large amount of data, we have to analyze the number of hours the players play on the server, which is numerical. As such, KNN regression would be a suitable method, especially knowing that the relationship between the number of hours played versus the players' characteristics may not be linear.

- We need to assume that the distance between points must represent similarity. KNN would look at the most similar players based on their characteristics (experience, subscribe, gender), and calculate the average of their play hours. A limitation for KNN regression is that it may not predict well beyond the range of values in our training data. 

- The data will be split into the training (75%) and test set (25%). 

- The KNN regression model will be evaluated by testing different values of k through cross-validation. The K value with the lowest RMSE would be used for the final model.
